### Task 4

```python
With the help of glob, use rename_my_file() to rename all the .txt files in the folder.

Note that you might have to adjust your function rename_my_file() to accommodate the already renamed files.
```

```python
import glob, os
```

```python
months_dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

def rename_my_file(old_file_name):
    file_name = os.path.basename(old_file_name)
    
    parts = file_name.split('-')

    # Check if file has already been renamed
    if parts[0].isdigit():
        print(f'{file_name} has been renamed')
        return None

    # New naming convention for the text files
    month_alphabets, rest = file_name.split('-', 1)
    date, city_extension = rest.split('_', 1)
    
    month_num = months_dict(month_alphabets)

    new_file_name = f'{month_num}-{date}-{city_extension}'

    # Tell the function what to do
    return new_file_name
    

# List all the text files in the selected folder
selected_files = glob.glob(os.path.join('os-collaboration-exercise-data', '*.txt'))

for selected_file in selected_files:
    new_file_name = rename_my_file(selected_file)
    
    if new_file_name:
        new_file_path = os.path.join('os-collaboration-exercise-data', new_file_name)
        
        # Rename the file
        os.rename(selected_file, new_file_path)
        print(f'{selected_file} has been renamed to {new_file_path}')

```

```python
import glob, os, shutil
```

```python
original_folder = 'os-collaboration-exercise-data'
city_folders = [f for f in glob.glob('*') if os.path.isdir(f)] # subdirectories within the folder i am in

city_related_files = glob.glob(os.path.join(original_folder, '*'))

for file_path in city_related_files:
    file_name = os.path.basename(file_path)
    
    for city_folder in city_folders:
        if city_folder.lower() in file_name: # To match the city related files to their corresponding folders
            final_folder = os.path.join(city_folder, file_name)

            shutil.move(city_related_files, final_folder) # To move the files accordingly
            print(f'{file_name} moved to {final_folder}')
            
            break # To increase efficiency and ensure that each file is only moved to their corresponding folder
```
